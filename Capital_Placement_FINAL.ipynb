```python
#PDF EXTRACTOR

import pdfplumber

# Function to extract text from a PDF file using pdfplumber
def extract_text_from_pdf(pdf_file):
    text = ""
    with pdfplumber.open(pdf_file) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Function to extract education and skills sections
def extract_education(text):
    lines = text.split('\n')
    headers = ["Contact Information","Objective","Summary","Experience","Education","Certifications","Projects","Awards","Achievements","Publications","Languages","Interests","References","Activities","Additional Information"]
    
    education_section = []
    
    in_education_section = False
    flag = False
    
    for line in lines:
        line = line.strip()
        
        if "Education" in line:
            in_education_section = True
            flag = True
        else:
            for header in headers:
                if header in line and flag:
                    in_education_section = False
                    return education_section
        
        if in_education_section and line:
            education_section.append(line)
    
    return education_section

def extract_skills(text):
    lines = text.split('\n')
    headers = ["Contact Information","Objective","Summary","Experience","Education","Certifications","Projects","Awards","Achievements","Publications","Languages","Interests","References","Activities","Additional Information"]
    
    skills_section = []
    
    in_skills_section = False
    flag = False
    
    for line in lines:
        line = line.strip()
        
        if "Skills" in line:
            in_skills_section = True
            flag = True
        else:
            for header in headers:
                if header in line and flag:
                    in_skills_section = False
                    return skills_section
        
        if in_skills_section and line:
            skills_section.append(line)
    
    return skills_section

# Replace 'resume.pdf' with the path to your PDF file
pdf_file = 'C:/Users/SATYESH/Downloads/resumedataset/data/data/ACCOUNTANT/11163645.pdf'

# Extract text from the PDF using pdfplumber
resume_text = extract_text_from_pdf(pdf_file)

# Extract education and skills sections
education = extract_education(resume_text)
skills = extract_skills(resume_text)

# Print the extracted sections
print("Education Section:")
for line in education:
    print(line)

print("\nSkills Section:")
for line in skills:
    print(line)
```

```python
from transformers import DistilBertTokenizer, DistilBertModel
import torch
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load the DistilBERT tokenizer and model
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
model = DistilBertModel.from_pretrained("distilbert-base-uncased")

# Create a list to store CVs
cvs = []
current_cv = []

# Loop through the lines in education and skills
for line in education + skills:
    if line.startswith("Education:") or line.startswith("Skills:"):
        # If a new section begins, save the previous CV and start a new one
        if current_cv:
            cvs.append("\n".join(current_cv))
            current_cv = []
    current_cv.append(line)

# Add the last CV entry
if current_cv:
    cvs.append("\n".join(current_cv))

# Tokenize and embed CV details and job descriptions
cv_embeddings = []
job_description_embeddings = []

# Loop through each CV
for cv_text in cvs:
    inputs = tokenizer(cv_text, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        outputs = model(**inputs)
    embeddings = torch.mean(outputs.last_hidden_state, dim=1)
    cv_embeddings.append(embeddings[0].numpy())  # Reshape to 2D array

# Tokenize and embed job descriptions
job_description_texts = job_descriptions.tolist()
for job_desc_text in job_description_texts:
    inputs = tokenizer(job_desc_text, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        outputs = model(**inputs)
    embeddings = torch.mean(outputs.last_hidden_state, dim=1)
    job_description_embeddings.append(embeddings[0].numpy())  # Reshape to 2D array

# Calculate cosine similarity
similarity_scores = []

for jd_embedding in job_description_embeddings:
    scores = cosine_similarity([jd_embedding], cv_embeddings)  # Compute similarity for a single job description
    similarity_scores.append(scores[0])  # Append the result

# Now, similarity_scores will contain cosine similarity scores between each job description and all CVs.
```

```python
def job_id_extractor_all(path_to_resume,path_to_id):
    
    import os
    import pandas as pd

# Define the path to the parent folder containing 24 subfolders with resumes
    parent_folder_path = path_to_resume

# Initialize an empty list to store job categories
    all_job_categories = []

# Define the path to the CSV file
    csv_file_path = path_to_id

# Read the CSV file into a DataFrame
    df = pd.read_csv(csv_file_path)

# Function to extract job ID and category from a PDF file
    def extract_job_info(pdf_file_path):
        job_id = os.path.splitext(os.path.basename(pdf_file_path))[0]  # Extract the job ID from the PDF filename
        job_id = int(job_id)  # Convert the job ID to an integer
        category = df[df['ID'] == job_id]['Category'].values[0]  # Find the corresponding job category
        return job_id, category

# Iterate through all subdirectories and their files
    for root, dirs, files in os.walk(parent_folder_path):
        for pdf_file in files:
            if pdf_file.endswith(".pdf"):
                pdf_path = os.path.join(root, pdf_file)

            # Extract job ID and category
                job_id, category = extract_job_info(pdf_path)

            # Append to the list of job categories
                all_job_categories.append((job_id, category))

# Now, all_job_categories contains tuples with job IDs and their corresponding categories
# for all resumes in the 24 subfolders.
    print(all_job_categories)
```

```python
# Import required libraries
from transformers import DistilBertTokenizer, DistilBertModel
import torch
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load the DistilBERT tokenizer and model
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
model = DistilBertModel.from_pretrained("distilbert-base-uncased")

# Assuming you have all_resume_details as a dictionary with job IDs and CV details
# and job_descriptions as a Series of job descriptions

# Create a dictionary to store the top 5 CVs for each job description
top_cv_for_jobs = {}

# Tokenize and embed job descriptions
job_description_embeddings = []

job_description_texts = job_descriptions.tolist()
for job_desc_text in job_description_texts:
    inputs = tokenizer(job_desc_text, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        outputs = model(**inputs)
    embeddings = torch.mean(outputs.last_hidden_state, dim=1)
    job_description_embeddings.append(embeddings[0].numpy())  # Reshape to 2D array

# Calculate cosine similarity for each job description
for index, jd_embedding in enumerate(job_description_embeddings):
    similarity_scores = []

    for job_id, cv_details in all_resume_details.items():
        # Combine education and skills details for each CV
        cv_text = "\n".join(cv_details.get("Education Section", []) + cv_details.get("Skills Section", []))

        # Tokenize and embed CV details
        inputs = tokenizer(cv_text, return_tensors="pt", padding=True, truncation=True)
        with torch.no_grad():
            outputs = model(**inputs)
        cv_embedding = torch.mean(outputs.last_hidden_state, dim=1)

        # Compute cosine similarity between job description and CV
        similarity = cosine_similarity([jd_embedding], [cv_embedding[0].numpy()])[0][0]

        similarity_scores.append((job_id, similarity))

    # Sort the CVs by similarity score in descending order
    similarity_scores.sort(key=lambda x: x[1], reverse=True)

    # Store the top 5 CVs for this job description
    top_cv_for_jobs[f"Job_{index + 1}"] = similarity_scores[:5]

# Print the top 5 CVs for each job description
for job, top_cvs in top_cv_for_jobs.items():
    print(f"Top 5 CVs for {job}:")
    for cv_id, score in top_cvs:
        print(f"Job ID: {cv_id}, Similarity Score: {score}")
    print()
```
